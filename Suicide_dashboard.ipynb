```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
import plotly.express as px
import pycountry


import panel as pn
pn.extension('tabulator')
pn.extension('plotly')
```

## Clean Data

```python
def clean_data(df1, df2):
    df1 = df1.drop_duplicates()
    df1 = df1.drop(columns=['Indicator Name','Indicator Code'])
    df1 = df1.fillna('0')
    
    df2 = df2.drop_duplicates()
    df2 = df2.fillna(' ')
    df2['IncomeGroup'] = df2['IncomeGroup'].replace(['Low & middle income', 'Lower middle income'], 'Lower middle income', regex=True)
    
    merged_data = pd.merge(df1, df2, on= 'Country Code', how = 'left')
    data = merged_data.reindex(columns=['Country Name', 'Country Code',
       'Region', 'IncomeGroup','2000', '2001', '2002', '2003',
       '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019'])
    
    year = data.columns.values[4:]
    data[year] = data[year].astype(float)
    
    return data
```

```python
data = clean_data(df1, df2)
```

## Create Individual Graphs

```python
slider = pn.widgets.IntSlider(name='Year slider ', start=2000, end=2019, step=1, value=2000)
```

### Bar Graph

```python
databar_clean = data.drop(columns=['Country Name', 'Country Code', 'Region'])
databar_clean = databar_clean.drop(data[data['IncomeGroup'] == ' '].index)
```

```python
databar_clean
```

```python
def plot_bar(Year):
    
    data_bar = databar_clean.groupby('IncomeGroup')[str(Year)].mean().reset_index()
    data_bar = data_bar.reindex(index = [0, 4, 3, 2, 1])
    
    bar_colors = ['#50180b', '#81120b', '#aa864d','#e4cda9', '#efe8dd']
    
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.bar(data_bar.iloc[:,0], data_bar.iloc[:,1], color=bar_colors)
    ax.tick_params(axis='x', rotation=45)
    
    plt.close(fig)
    
    #plt.bar(data_bar.iloc[:,0], data_bar.iloc[:,1], color=bar_colors)
    #plt.xticks(rotation = 45)
    #plt.show()
    
    return pn.pane.Matplotlib(fig, width=600, height=400)
```

```python
widgets.interact(plot_bar, Year = (2000, 2019, 1))
```

### Line Graph

```python
dataline_clean = data.drop(columns=['Country Code','Region','IncomeGroup'])
```

```python
dataline_clean
```

```python
def plot_line(Year):
    
    melted_data_line = dataline_clean.melt(id_vars='Country Name', var_name='Year', value_name='Value')
    data_line = melted_data_line.groupby(['Year'])['Value'].mean().reset_index()

    till_year = data_line[data_line['Year'] == str(Year)].index[0] + 1
    data_line = data_line.iloc[0:till_year]
    
    fig, ax = plt.subplots()
    ax.plot(data_line.iloc[:,0], data_line.iloc[:,1], color= 'maroon', marker= ".")
    ax.tick_params(axis='x', rotation=45)
    
    plt.close(fig)
    
    #plt.plot(data_line.iloc[:,0], data_line.iloc[:,1], color= 'maroon', marker= ".")
    #plt.xticks(rotation = 45)
    #plt.show()
    return pn.pane.Matplotlib(fig, width=600, height=400)
```

```python
widgets.interact(plot_line, Year = (2000, 2019, 1))
```

### Table

```python
datatable_clean = data.drop(columns=['Country Code'])
datatable_clean = datatable_clean.melt(id_vars=['Country Name', 'Region', 'IncomeGroup'], var_name='Year', value_name='Value')
datatable_clean = datatable_clean.drop(datatable_clean[datatable_clean['IncomeGroup'] == ' '].index)
```

```python
datatable_clean
```

```python
def create_table(Year):
    
    data_table = datatable_clean[datatable_clean['Year'] == str(Year)].sort_values(by='Value', ascending=False).reset_index()
    data_table = data_table.drop(columns=['index','Year'])
    data_table = data_table.rename(columns={'Value':'Rate'})
    data_table = data_table[0:11]
    
    table_widget = pn.widgets.Tabulator(data_table, theme='simple')
    
    return table_widget
```

```python
widgets.interact(create_table, Year = (2000, 2019, 1))
```

### Map

```python
def alpha3_code():
    country_code = list(pycountry.countries)
    alpha3 = []

    for i in range(len(country_code)):

        alpha3.append(country_code[i].alpha_3)

    data_alpha3 = pd.DataFrame(alpha3)
    data_alpha3 = data_alpha3.rename(columns={0: 'Country Code'})
    
    return data_alpha3
```

```python
data_alpha3 = alpha3_code()
datamap_merge = pd.merge(data, data_alpha3, on= 'Country Code', how = 'inner')
datamap_clean = datamap_merge.drop(columns=['Region','IncomeGroup'])
```

```python
datamap_clean
```

```python
def create_map(Year):
    
    data_map = datamap_clean[['Country Name','Country Code',str(Year)]]
    color_scale = [(0, '#e9e2c7'), (0.5, '#a46e51'), (1, '#4f0205')]

    fig = px.choropleth(
        data_map,
        locations='Country Code',
        color=str(Year), 
        hover_name='Country Name', 
        projection='natural earth',
        color_continuous_scale=color_scale
    )

    return pn.pane.Plotly(fig)
```

```python
widgets.interact(create_map, Year = (2000, 2019, 1))
```

### Create Dashboard

```python
bar_pane = pn.interact(plot_bar, Year=slider)
line_pane = pn.interact(plot_line, Year=slider)
table_pane = pn.interact(create_table, Year=slider)
map_pane = pn.interact(create_map, Year=slider)
```

```python
markdown_text = """
# Suicide Mortality Rate 
## (per 100,000 populations)

Welcome to the "Exploring Global Suicide Rates" dashboard. This interactive dashboard provides valuable insights into the global trends of suicide rates across various countries and years. Suicide is a critical public health issue that affects individuals and communities worldwide. This dashboard aims to shed light on the patterns, factors, and potential correlations associated with suicide rates.

### Interactive Features
- **Map:** showing the suicide rate by countries
- **Bar Chart:** showing the average suicide rate categorized by income groups
- **Line Chart:** showing world suicide rate trend
- **Table:** showing the top 10 suicide rates by country

[Link to resource](https://data.worldbank.org/indicator/SH.STA.SUIC.P5?end=2019&name_desc=false&start=2019&type=shaded&view=map&year=2018)

### Setting
"""
```

```python
template = pn.template.FastListTemplate(
    title = 'Global Suicide Rates Dashboard',
    sidebar = [pn.pane.Markdown(markdown_text),
               slider],
    main = [pn.Row(pn.Column(map_pane, width=600, height=400), pn.Column(bar_pane, width=600, height=400)),
            pn.Row(pn.Column(line_pane, width=600, height=400), pn.Column(table_pane, width=600, height=400))],
    accent_base_color="#A01346",
    header_background="#50180b",
)

template.show()
```
